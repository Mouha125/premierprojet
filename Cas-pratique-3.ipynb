```python
import pandas as pd
import numpy as np
df=pd.ExcelFile("Del_Annuel.xlsx", engine='openpyxl')
R=df.parse("PV")
w=R.loc[9]
R
```

```python
B=R.index[pd.isnull(R).any(axis=1)]
R.drop(B, inplace=True)
#R.dropna()
R
```

```python
R.columns=w
bon_R=R.rename(columns={'RESULTAT': 'Total_Crédits'})
bdf=bon_R.columns.fillna("Décision")
bon_R.columns=bdf
bon_R
```

```python
l=bon_R.reset_index()
del l['index']
l
```

```python
tri_l=l.sort_values(by = ["Total_Crédits","Nom","Prénom"],
ascending = [False,True,True])
tri_l
```

```python
Ps=tri_l.Décision.value_counts()
Ps
print("Le nombres d'étudiants en passage définitif est: ", Ps[0])
print("Le nombres d'étudiants en passage conditionnel est: ", Ps[2])
print("Le nombres d'étudiants qui redoublent est: ", Ps[1])
print("Le nombres d'étudiants exclu est: ", Ps[3] )
```

```python
df=pd.DataFrame(Ps)
df.columns=["Nbr_d'étudiants"]
o=df.values.sum()
df.loc["Effectif_Total"]=[o]
j=[Ps[0]/o*100,Ps[1]/o*100,Ps[2]/o*100,Ps[3]/o*100]
m=Ps[0]/o*100+Ps[1]/o*100+Ps[2]/o*100+Ps[3]/o*100
j.append(m)
df=df.assign(Pourcentage=j)
df
```

```python
g=[Ps[0]/o*100,Ps[1]/o*100,Ps[2]/o*100,Ps[3]/o*100]
g
```

```python
import matplotlib.pyplot as plt
labels= 'Passage-Définitif', 'Redouble', 'Passage-Conditionnel', 'Exclu'
serie=g
separation = (0, 0, 0, 0)
fig, ax= plt.subplots()
ax.pie(serie, explode=separation, labels=labels, autopct='%1.0f%%', shadow= False, startangle=90)
ax.axis('equal')
plt.show()
```

```python
writer = pd.ExcelWriter("délibération.xlsx", engine="xlsxwriter")
df.to_excel(writer, "Résultats Généraux")
writer.save()
```
